```python
import torch
from ct_pretraining.schedulers.decaying_sine import DecayingSineLR
from matplotlib import pyplot as plt
from torch import nn
```

```python
optimizer = torch.optim.Adam([torch.zeros(1)], lr=0.1)
scheduler = DecayingSineLR(optimizer, 1e-6, 1e-3, 5000 // 4, 0.5)
```

```python
lrs = []

for _ in range(5000):
    lrs.append(scheduler.get_last_lr()[0])
    scheduler.step()

lrs[:5]
```

```python
plt.plot(lrs)
plt.show()
```
